```python
import pandas as pd
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
sum = pd.read_csv("Sum.txt", sep="\t", header=None)
sum.columns = ["worst", "total", "initial worst", "initial total"]

sum
```

```python
worst=sum[sum.filter(regex='worst').columns]
display(worst)

```

```python
plt.figure()
worst.plot.line()
plt.tight_layout()
plt.savefig("worst_line.png", dpi=300)
```

```python
plt.figure()
worst.plot.bar()
plt.tight_layout()
plt.savefig("worst_bar.png", dpi=300)
```

```python
total=sum[sum.filter(regex='total').columns]
display(total)
```

```python
plt.figure()
total.plot.line()
plt.tight_layout()
plt.savefig("total_line.png", dpi=300)
```

```python
plt.figure()
total.plot.bar()
plt.tight_layout()
plt.savefig("total_bar.png", dpi=300)
```

```python
ax=sns.boxplot(data=total)
ax.set( ylabel='minutes')
plt.savefig("total_box.png", dpi=300)

plt.show()

ax=sns.boxplot(data=worst)
ax.set( ylabel='minutes')
plt.savefig("worst_box.png", dpi=300)

plt.show()
```

```python
ave = pd.read_csv("Average.txt", sep="\t", header=None)
ave.columns = ["worst", "total", "initial worst", "initial total"]

ave
```

```python
ave_worst=ave[ave.filter(regex='worst').columns]
display(ave_worst)
ave_total=ave[ave.filter(regex='total').columns]
display(ave_total)

```

```python
ax=sns.boxplot(data=ave_total)
ax.set( ylabel='minutes')
plt.savefig("ave_total_box.png", dpi=300)

plt.show()

ax=sns.boxplot(data=ave_worst)
ax.set( ylabel='minutes')
plt.savefig("ave_worst_box.png", dpi=300)

plt.show()
```

```python
compare_total=pd.concat([total, ave_total], axis=0)
compare_total['Type']=0
compare_total.iloc[0:10, 2]='Sum'
compare_total.iloc[10:, 2]='Average'
compare_total['t_fixed']=0
compare_total['it_fixed']=0
compare_total.iloc[:, 3]=compare_total['total']*60
compare_total.iloc[:, 4]=compare_total['initial total']*60
compare_total.iloc[:10, 3]=compare_total.iloc[:10, 0]
compare_total.iloc[:10, 4]=compare_total.iloc[:10, 1]
compare_total
```

```python
compare_worst=pd.concat([worst, ave_worst], axis=0)
test=pd.melt(compare_worst)
compare_worst['Type']=0
compare_worst.iloc[0:10, 2]='Sum'
compare_worst.iloc[10:, 2]='Average'
compare_worst
```

```python
test=pd.concat([worst, ave_worst], axis=0)
test=pd.melt(test)
test['type']=0
test.iloc[0:10, 2]='Sum'
test.iloc[10:20, 2]='Average'
test.iloc[20:30, 2]='Sum'
test.iloc[30:40, 2]='Average'
test
```

```python
sns.boxplot(data=test, x='type', y='value', hue='variable')
plt.savefig("compare_worst_box.png", dpi=300)
```

```python
test_total=compare_total.iloc[:, 3:5]
test_total.rename(columns = {'t_fixed':'total'}, inplace = True)
test_total.rename(columns = {'it_fixed':'initial total'}, inplace = True)
test_total=pd.melt(test_total)
test_total['type']=0
test_total.iloc[0:10, 2]='Sum'
test_total.iloc[10:20, 2]='Average'
test_total.iloc[20:30, 2]='Sum'
test_total.iloc[30:40, 2]='Average'
test_total
```

```python
final=pd.concat([test, test_total], axis=0)
final=final[~final['variable'].str.contains('initial', na=False)]

final
```
